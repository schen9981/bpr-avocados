```python
import pandas
avocados = pandas.read_csv(
    'C:\\Users\\kycqi\\Documents\\Comp Sci\\BPR\\avocado.csv',
    index_col=0,
)
import seaborn
seaborn.set_style('white')
```

```python
avocados['Date'] = pandas.to_datetime(avocados['Date'])
avocados.head()
```

```python
mask = avocados['type']=='conventional'
graph = seaborn.catplot('AveragePrice','region',data=avocados[mask],
                   hue='year',
                   height=8,
                   aspect=0.6,
                   palette='BuGn',
              )
```

```python
order = (
    avocados[mask & (avocados['year']==2018)]
    .groupby('region')['AveragePrice']
    .mean()
    .sort_values()
    .index
)
```

```python
graph = seaborn.catplot('AveragePrice','region',data=avocados[mask],
                   hue='year',
                   height=8,
                   aspect=0.6,
                   palette='BuGn',
                   order=order,
              )
```
